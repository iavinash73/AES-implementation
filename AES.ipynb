```python
# Splits input plaintext to blocks of 16 bytes (4x4 grid) and also converts it into ascii
# Takes in a string and returns array of 4x4 grids of ascii values

def break_in_grids_of_16(s):
    all = []
    for i in range(len(s)//16):
        b = s[i*16: i*16 + 16]
        grid = [[], [], [], []]
        for i in range(4):
            for j in range(4):
                grid[i].append(ord(b[i + j*4]))
        all.append(grid)
    return all
```

```python
# Shift-Row - Takes in a array and makes left circular shifts.


def rotate_row_left(row, n=1):
    return row[n:] + row[:n]
```

```python
# Key Expansion - Takes 4x4 key grid and generates 4 new words per round (i.e, 16 new bytes = 128 bits per round)


def expand_key(key,rounds):
    
    # Round constant for key generation
    rcon = [[1, 0, 0, 0]]
    for _ in range(1, rounds):
        rcon.append([rcon[-1][0]*2, 0, 0, 0])
        if rcon[-1][0] > 0x80:
            rcon[-1][0] ^= 0x11b
            
    key_grid = key
    for round in range(rounds):
        # Takes the last column (Last word)
        last_column = [row[-1] for row in key_grid]
        # Perform Shift row
        last_column_rotate_step = rotate_row_left(last_column)
        # S-box 
        last_column_sbox_step = [lookup(b) for b in last_column_rotate_step]
        # XOR the result with round constant
        last_column_rcon_step = [last_column_sbox_step[i]
                                 ^ rcon[round][i] for i in range(len(last_column_rotate_step))]

        # XOR with the generated word and a word from previous key
        for r in range(4):
            key_grid[r] += bytes([last_column_rcon_step[r]
                                  ^ key_grid[r][round*4]])
            
        # Continue the chain for 3 more columns
        for i in range(len(key_grid)):
            for j in range(1, 4):
                key_grid[i] += bytes([key_grid[i][round*4+j]
                                      ^ key_grid[i][round*4+j+3]])
                

    return key_grid
```

```python
# Extract the Sub-key for each round.

def extract_key_for_round(expanded_key, round):
  return [row[round*4: round*4 + 4] for row in expanded_key]
```

```python
# Add round key - takes in block grid and key grid and returns added block grid

def add_sub_key(block_grid, key_grid):
    r = []

    for i in range(4):
        r.append([])
        for j in range(4):
            r[-1].append(block_grid[i][j] ^ key_grid[i][j])
    return r
```

```python
# Mix column - takes in plaintext's grid and returns resultant grid


# we are multiplying the grid against this matrix
# ╔              ╗
# ║  2  3  1  1  ║
# ║  1  2  3  1  ║
# ║  1  1  2  3  ║
# ║  3  1  1  2  ║
# ╚              ╝


def multiply_by_2(v):
    #multiply by 2
    s = v << 1
    #only 8 LSB are retained
    s &= 0xff
    #if v > 128 (i.e, 8th bit is set as 1)
    if (v & 128) != 0:
        s = s ^ 0x1b
    return s


def multiply_by_3(v):
    return multiply_by_2(v) ^ v


def mix_columns(grid):
    new_grid = [[], [], [], []]
    for i in range(4):
        #take first column
        col = [grid[j][i] for j in range(4)]
        col = mix_column(col)
        #store in new grid
        for i in range(4):
            new_grid[i].append(col[i])
    return new_grid

#function for each column
def mix_column(column):
    r = [
        multiply_by_2(column[0]) ^ multiply_by_3(column[1]) ^ column[2] ^ column[3],
        multiply_by_2(column[1]) ^ multiply_by_3(column[2]) ^ column[3] ^ column[0],
        multiply_by_2(column[2]) ^ multiply_by_3(column[3]) ^ column[0] ^ column[1],
        multiply_by_2(column[3]) ^ multiply_by_3(column[0]) ^ column[1] ^ column[2],
    ]
    return r
```

```python

def enc(data):

    grids = break_in_grids_of_16(data)
    print("Array of ASCII before encryption: ", grids)
    key=[[0x2b, 0x7e, 0x15, 0x16],
    [0x28, 0xae, 0xd2, 0xa6],
    [0xab, 0xf7, 0x97, 0x32],
    [0xa3, 0x45, 0x00, 0x98]]

    # Now we need to expand the key for the multiple rounds
    expanded_key = expand_key(key, 11)

    # INITIAL PERMUTAION (ADD ROUND KEY) - Apply the original key to the blocks before start the rounds
    temp_grids = []
    round_key = extract_key_for_round(expanded_key, 0)

    for grid in grids:
        temp_grids.append(add_sub_key(grid, round_key))

    grids = temp_grids

    # 10 Rounds for AES 128.
    for round in range(1, 10):
        temp_grids = []
        
        for grid in grids:
            # S-box
            sub_bytes_step = [[lookup(val) for val in row] for row in grid]
            # Shift-rows
            shift_rows_step = [rotate_row_left(
                sub_bytes_step[i], i) for i in range(4)]
            # Mix Columns
            mix_column_step = mix_columns(shift_rows_step)
            # Get Sub-key and add it
            round_key = extract_key_for_round(expanded_key, round)
            add_sub_key_step = add_sub_key(mix_column_step, round_key)
            temp_grids.append(add_sub_key_step)

        grids = temp_grids

    # A final round without the mix columns
    temp_grids = []
    round_key = extract_key_for_round(expanded_key, 10)

    for grid in grids:
        sub_bytes_step = [[lookup(val) for val in row] for row in grid]
        shift_rows_step = [rotate_row_left(
            sub_bytes_step[i], i) for i in range(4)]
        add_sub_key_step = add_sub_key(shift_rows_step, round_key)
        temp_grids.append(add_sub_key_step)

    grids = temp_grids
    print("Array of ASCII after encryption: ", grids)

    # Recriate the data into a single stream before returning
    res = '';
    
    for grid in grids:
        for column in range(4):
            for row in range(4):
                res+=bin(grid[row][column])

    return res
```

```python
def dec():

    grids = [[[10, 37, 176, 25], [200, 44, 188, 114], [218, 144, 204, 87], [66, 201, 126, 207]]]
    key = [[0x2b, 0x7e, 0x15, 0x16],
    [0x28, 0xae, 0xd2, 0xa6],
    [0xab, 0xf7, 0x97, 0x32],
    [0xa3, 0x45, 0x00, 0x98]]
    expanded_key = expand_key(key, 11)
    temp_grids = []
    round_key = extract_key_for_round(expanded_key, 10)

    # First we undo the final round
    temp_grids = []

    for grid in grids:

        add_sub_key_step = add_sub_key(grid, round_key)
        shift_rows_step = [rotate_row_left(
            add_sub_key_step[i], -1 * i) for i in range(4)]
        sub_bytes_step = [[reverse_lookup(val) for val in row]
                          for row in shift_rows_step]
        temp_grids.append(sub_bytes_step)

    grids = temp_grids

    for round in range(9, 0, -1):
        temp_grids = []

        for grid in grids:
            round_key = extract_key_for_round(expanded_key, round)
            add_sub_key_step = add_sub_key(grid, round_key)

            # Doing the mix columns three times is equal to using the reverse matrix
            mix_column_step = mix_columns(add_sub_key_step)
            mix_column_step = mix_columns(mix_column_step)
            mix_column_step = mix_columns(mix_column_step)
            shift_rows_step = [rotate_row_left(
                mix_column_step[i], -1 * i) for i in range(4)]
            sub_bytes_step = [
                [reverse_lookup(val) for val in row] for row in shift_rows_step]
            temp_grids.append(sub_bytes_step)

        grids = temp_grids
        temp_grids = []

    # Reversing the first add sub key
    round_key = extract_key_for_round(expanded_key, 0)

    for grid in grids:
        temp_grids.append(add_sub_key(grid, round_key))

    grids = temp_grids

    # Just transform the grids back to bytes
    int_stream = []
    for grid in grids:
        for column in range(4):
            for row in range(4):
                int_stream.append(grid[row][column])

    return bytes(int_stream)
```
